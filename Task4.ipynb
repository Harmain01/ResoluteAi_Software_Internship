```python
import cv2
import numpy as np

bgr= cv2.imread(r"C:\Users\ASUS\Documents\Task 4\mainimg.jpg")
gray= cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)

template= cv2.imread(r"C:\Users\ASUS\Documents\Task 4\key.jpg",0)
w, h= template.shape[::-1]

res= cv2.matchTemplate(gray,template,cv2.TM_CCOEFF_NORMED)
threshold= 0.8
loc= np.where(res>+threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(bgr,pt,(pt[0]+w, pt[1]+h), (255,0,255), 2)    #(255,0,255)= purp;e 
    #2-> line width
cv2.imshow('detected', bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Foreground Extraction

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread(r"C:\Users\ASUS\Documents\Task 4\me.jpg")
mask = np.zeros(img.shape[:2], np.uint8)

back= np.zeros((1,65), np.float64)
fore= np.zeros((1,65), np.float64)
rect=(450,345,175,250)  #(x,y,width,height)
cv2.grabCut(img, mask,rect, back, fore, 5, cv2.GC_INIT_WITH_RECT)
mask2= np.where((mask==2)|(mask==0),0,1).astype('uint8')
img=img*mask2[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()
```

Corner Detection

```python
import cv2
import numpy as np
img= cv2.imread(r"C:\Users\ASUS\Documents\Task 4\corner_detection.jpg")
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray= np.float32(gray) 
#cv2.imshow('gray',gray)

corners= cv2.goodFeaturesToTrack(gray,60,0.01,10)
corners=np.int0(corners)

for corners in corners:
    x,y= corners.ravel()
    cv2.circle(img, (x,y), 3,255,-1)    #cv2.circle(img, center, radius, color, thickness)

cv2.imshow('corners',img)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

Feature Matching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1=cv2.imread(r"C:\Users\ASUS\Documents\Task 4\feature_matching_img.jpg",0)
img2=cv2.imread(r"C:\Users\ASUS\Documents\Task 4\feature-matching-template.jpg",0)

orb= cv2.ORB_create()

kp1, des1= orb.detectAndCompute(img1, None)
kp2, des2= orb.detectAndCompute(img2, None)

bf= cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck= True)

matches= bf.match(des1,des2)
matches= sorted(matches, key= lambda x:x.distance)

img3= cv2.drawMatches(img1,kp1,img2,kp2, matches[:12], None, flags=2)
plt.imshow(img3)
plt.show()
```

OBJECT DETECTION USING HAAR CASCADE 

```python
import cv2
import numpy as np

face_cascade=cv2.CascadeClassifier(r"C:\Users\ASUS\Documents\Task 4\haarcascade_frontalface_default.xml")
eye_cascade=cv2.CascadeClassifier(r"C:\Users\ASUS\Documents\Task 4\haarcascade_eye.xml")

cap=cv2.VideoCapture(0)

while True:
    ret, img= cap.read()
    gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    for [x,y,w,h] in faces:
        cv2.rectangle(img, (x,y),(x+w,y+h),(0,0,0),2)
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=img[y:y+h,x:x+w]
        eyes=eye_cascade.detectMultiScale(roi_gray)
        for [ex,ey,ew,eh] in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
    cv2.imshow('img',img)
    if cv2.waitKey(5) & 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
